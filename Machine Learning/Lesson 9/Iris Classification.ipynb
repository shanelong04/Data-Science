ENSEMBLE LEARNING 

```python
import pandas as pd 
from sklearn.datasets import load_iris 
```

```python
df = load_iris()
df.data.shape
```

```python
df.feature_names
df.target_names
X = df.data
y = df.target 
seed = 8 
tree = 30
```

```python
from sklearn.ensemble import AdaBoostClassifier 
from sklearn import model_selection 
```

```python
kfold = model_selection.KFold(n_splits = 10)
model = AdaBoostClassifier(n_estimators = tree, random_state = seed)
results = model_selection.cross_val_score(model,X,y,cv=kfold)
print(results.mean())
```

```python
model.fit(X,y)
print(model.predict([[2,1,2,23]]))
```
