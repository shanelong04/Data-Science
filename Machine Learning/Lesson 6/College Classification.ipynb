```python
import pandas as pd
```

```python
df = pd.read_csv('college.csv')
df.tail()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X = df.drop(['Private'],axis = 1)
y = df['Private']
label = LabelEncoder()
y = label.fit_transform(y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 100)
```

```python
from sklearn.svm import SVC
svc_model = SVC()
svc_model.fit(X_train,y_train)
predictions = svc_model.predict(X_test)
score = accuracy_score(y_test,predictions)
score
```

```python
# Using standard scaler to improve the performance/accuracy 
scaler = StandardScaler()
X = scaler.fit_transform(X)
X
```

```python
from sklearn.svm import SVC
svc_model = SVC()
svc_model.fit(X_train,y_train)
predictions = svc_model.predict(X_test)
score = accuracy_score(y_test,predictions)
score
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': [0.1,1,10,100], 'gamma': [1,0.1,0.01,0.001]}
# Tim cac hyperparameters toi uu de cai thien hieu suat/do chinh xac cua du doan bang cach 
# tao mot luoi sieu tham so va thu tat ca cac ket hop cua chung xem cai nao la toi uu nhat
grid = GridSearchCV(SVC(),param_grid,refit = True,verbose = 2)
grid.fit(X_train,y_train)
```

```python
predictions = grid.predict(X_test)
print(accuracy_score(y_test,predictions))
```
