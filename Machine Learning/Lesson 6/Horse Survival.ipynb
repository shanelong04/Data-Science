```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df = pd.read_csv('horse.csv')
df
```

```python
df.columns
```

```python
target = df['outcome']
target.unique()
```

```python
df = df.drop(['outcome'],axis = 1)
```

```python
category_variables = ['surgery','age','temp_of_extremities','peripheral_pulse','mucous_membrane','abdomen',
                      'capillary_refill_time','pain','peristalsis', 'abdominal_distention','nasogastric_tube','nasogastric_reflux','rectal_exam_feces','abdomo_appearance',
                      'surgical_lesion','cp_data']
```

```python
for category in category_variables:
    df[category] = pd.get_dummies(df[category])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
X,y = df,target
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 1)
```

```python
from sklearn.tree import DecisionTreeClassifier 
print(X_train.shape)
```

```python
from sklearn.impute import SimpleImputer
#Ham thay the missing values khac - thay vi su dung ham fillna
imp = SimpleImputer(missing_values = np.nan,strategy = "most_frequent")
X_train = imp.fit_transform(X_train)
X_test = imp.fit_transform(X_test)
```

```python
classifier = DecisionTreeClassifier()
classifier.fit(X_train,y_train)
```

```python
predictions = classifier.predict(X_test)
from sklearn.metrics import accuracy_score
score = accuracy_score(predictions,y_test)
score
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier = RandomForestClassifier()
classifier.fit(X_train,y_train)
```

```python
predictions = classifier.predict(X_test)
score = accuracy_score(predictions,y_test)
score
```
