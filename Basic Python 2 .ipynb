```python
def tra_ve_list(n):
    a = [ i for i in range(n)]
    b = [ i*i for i in range(n)]
    return a,b
try:
    n = int(input())
    if n<0:
        print('Nhap vao so tu nhien n!')
    else:
        a,b = tra_ve_list(n)
        print(*a)
        print(*b)
except:
    print('Dinh dang dau vao khong hop le')
```

```python
def tbc(x):
    sum(x)
    len(x)
    return sum(x)/len(x)  
try:
    lst = list(map(float,input().split()))
    if len(lst)==0:
        print('Danh sach rong!')
    else:
        tbc(lst)
        print('Gia tri trung binh cong cua day la {}'.format(tbc(lst)))
except:
    print('Dinh dang dau vao khong hop le!')
```

```python
def phantumax(ds):
    giatri = max(ds)
    soluong = ds.count(giatri)
    vitri = [vt for vt in range(0,len(ds)) if ds[vt]==giatri]
    return giatri,soluong,vitri
try:
    lst = list(map(float,input().split()))
    if len(lst)==0:
        print('Danh sach rong!')
    else:
        giatri, soluong, vitri = phantumax(lst)
        print(giatri)
        print(soluong)
        print(*vitri)
except:
    print(' Vui long nhap cac phan tu la so thuc!')
```
