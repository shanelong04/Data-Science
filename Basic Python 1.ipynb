```python
# Tính tích của tổng các số chẵn và tổng các số lẻ của 1 stn:
try:
    n = int(input('Nhap vao so tu nhien n: '))
    if n<0:
        print('Vui long nhap so tu nhien')
    else:
        odd=0
        even=0
        while n > 0:
            chuSo = n % 10
            n = n // 10  
            if chuSo%2==0:
                even+=chuSo
            else:
                odd+=chuSo
        multi=odd*even
        print('Tich cua tong chu so chan va tong chu so le la {}'.format(multi))
except:
    print('Dinh dang dau vao khong hop le!')
```

```python
# *args được sử dụng khi kb nhập bn tham số đầu vào
def tinh_tong(*args):
   tong = 0
   for so in args:
       tong += so
   return tong
print("1 + 2 + 3 = {}".format(tinh_tong(1, 2, 3)))
print("1 + 2 + 3 + 4 + 5 = {}".format(tinh_tong(1, 2, 3, 4, 5)))
```

```python
# Máy tính đơn giản :
def maytinh(a,x,b):
    if x=='+':
        y = a+b
    elif x=='-':
        y = a-b
    elif x=='*':
        y = a*b
    else:
        y = a/b
    return y
try:
    a,x,b = input('Nhap vao so thu nhat, phep tinh va so thu 2: ').split()
    a,b = int(a),int(b)
    if x=='/' and b==0:
        print('So chia phai khac khong!')
    else:
        maytinh(a,x,b)
        print('{} {} {} = {}'.format(a,x,b,maytinh(a,x,b)))
except:
    print('Dinh dang dau vao khong hop le!')
```
