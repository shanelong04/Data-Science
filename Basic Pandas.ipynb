DATA PREPROCESSING WITH PANDAS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
df = pd.read_csv("SalaryGender.csv", delimiter = ',') #Import data
df.to_csv("filename.csv") #Load data to csv file
```

```python
#DATA EXPLORATION TECHNIQUES
df.info() #Read information
type(df) #type of dataset
df['Salary'].dtype
df.describe()
df.groupby('Salary')['Age'].sum() #Group
df[1:3] #slicing and indexing    
df.iloc[1:3,2:5] #fetching the data from row 1:2 and column 2:4
df['Age'].unique() #extracting all unique values out of 'Age' column
#df['Age'].values() returns a numpy array with all the values
df.mean()
df.median()
df.mode(axis = 1)
#Plotting a heatmap with seaborn
correlations = df.corr()
sns.heatmap(data = correlations, square = True, cmap = "bwr")
plt.yticks(rotation=0)
plt.xticks(rotation=90)
df[['Salary','Gender','Age']].corr() #Find the correlation among variables
#DATA WRANGLING (method of manually converting or mapping data from one raw format into another format)
#it includes munging and data visualization
#missing value and outlier 
from sklearn.datasets import load_diabetes
dataset = load_diabetes()
dataset.data
dataset.target
dataset['feature_names']
#df = pd.DataFrame(data,columns) converts dataset into dataframes
dataset.isnull()
#DATA MANIPULATION (head,tail,values,concatenate,merge)
#TYPECASTING (string,int,float)








```
